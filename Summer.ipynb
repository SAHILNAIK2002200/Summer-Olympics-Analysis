# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("summer.csv")
df
```

### 1. In how many cities Summer Olympics is held so far?

```python
data=[]
for city in df["City"].unique():
  data.append((city,len(df[df["City"]==city])))
data=pd.DataFrame(data,columns=["City","Frequency"]).sort_values(by="Frequency",ascending=False)
data

```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data=[]
for sport in df["Sport"].unique():
  data.append((sport ,len(df[(df['Sport'] == sport) & (df['Medal'] == "Gold")])))
data=pd.DataFrame(data,columns=["Sport","no_of_gold_medals"]).sort_values(by="no_of_gold_medals",ascending=False).head()
data.plot(x="Sport",y="no_of_gold_medals",kind="bar",figsize=(7,7))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data=[]
for sport in df["Sport"].unique():
  data.append((sport,len(df[df["Sport"]==sport]))) 
data=pd.DataFrame(data,columns=["Sport","no_of_medals"]).sort_values(by="no_of_medals",ascending=False).head()  
data.plot(x="Sport",y="no_of_medals",kind="bar",figsize=(7,6))
```

### 4. Which player has won most number of medals? (Top 5)

```python
data=[]
for athlete in df["Athlete"].unique():
  data.append((athlete,len(df[df["Athlete"]==athlete]))) 
data=pd.DataFrame(data,columns=["Athlete","no_of_medals"]).sort_values(by="no_of_medals",ascending=False).head()  
data.plot(x="Athlete",y="no_of_medals",kind="bar",figsize=(7,7))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
data=[]
for athlete in df["Athlete"].unique():
  data.append((athlete ,len(df[(df['Athlete'] == athlete) & (df['Medal'] == "Gold")])))
data=pd.DataFrame(data,columns=["Athlete","no_of_gold_medals"]).sort_values(by="no_of_gold_medals",ascending=False).head()  
data.plot(x="Athlete",y="no_of_gold_medals",kind="bar",figsize=(7,7))
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data=[]
for event in df["Event"].unique():
  data.append((event,len(df[df["Event"]==event])))
data=pd.DataFrame(data,columns=["Event","no_of_players"]).sort_values(by="no_of_players",ascending=False).head()
data.plot(x="Event",y="no_of_players",kind="bar",figsize=(7,7))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data=[]
for sport in df["Sport"].unique():
  data.append((sport ,len(df[(df['Sport'] == sport) & (df['Medal'] == "Gold") & (df["Gender"]=="Women")])))
data=pd.DataFrame(data,columns=["Sport","Female_gold_medalist"]).sort_values(by="Female_gold_medalist",ascending=False).head()
data.plot(x="Sport",y="Female_gold_medalist",kind="bar",figsize=(7,7))
```
